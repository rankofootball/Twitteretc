```python
import time    
import urllib2  

#Downloading entire Web Document (Raw Page Content)
def download_page(url):
    try:
        req = urllib2.urlopen(url)
        respData = str(req.read())
        return respData
    except Exception as e:
        print(str(e))

def get_first_normal_link(s):
    link = "no_links"
    while True:
        start_link = s.find("<a href=\"/wiki/")
        start_par = s.find("<p")  # link should appear in main text paragraph
        end_par = s.find("</p>")  #( I have assumed it is marked by <p>)

        if start_link == -1:    #If no links are found then give an error!
            return link

        if start_par < end_par:
            s = str(s[start_par+1:])
        elif start_link > end_par:
            s = str(s[start_par+1:])
        else:
            link = str(s[start_link+1:start_link+20]) 
            if link.find(":") > -1 or link.find("class") > -1:       # discard images etc.
                start_quote = s.find('"', start_link)
                end_quote = s.find('"',start_quote+1)
                s = str(s[end_quote+1:])

            else:
                start_quote = s.find('"', start_link)
                end_quote = s.find('"',start_quote+1)
                link = str(s[start_quote+1:end_quote])
                return link

# crawling loop        
def crawl(starting_page):

    page = download_page(starting_page)
    link = 0 
#    counter = 0
    seen = []
    while link > -1:
        
        link = get_first_normal_link(page)
        print str(link[6:])
        if link.find("Philosophy") > -1:
            print "-- Philosophy article found --"
            break

        if link in seen:           # detected loop
            print "-- Philosophy article not found --"
            break

        page = download_page("https://en.wikipedia.org/"+link)
#        if counter > -1: counter += 1
        seen.append(link)
#        if counter > 100:
#            seen = link
#            counter = -1
        time.sleep(0.5)

#main    
crawl(" https://en.wikipedia.org/wiki/Special:Random")
```
